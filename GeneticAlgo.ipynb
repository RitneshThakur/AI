```python
import numpy as np
class GeneticAlgorithm:
def __init__(self, population_size, num_generations, mutation_rate, tournament_size):
self.population_size = population_size
self.num_generations = num_generations
self.mutation_rate = mutation_rate
self.tournament_size = tournament_size
self.population = self.initialize_population()
def fitness(self, x):
return abs(int(x**2 + 9*x + 18))
def decode(self, binary_string):
integer_part = int(binary_string[:3], 2)
fractional_part = int(binary_string[3:], 2) / (2**5)
return integer_part + fractional_part
def initialize_population(self):
return [''.join(np.random.choice(['0', '1'], 8)) for _ in range(self.population_size)]
def tournament_selection(self):
tournament = np.random.choice(self.population, self.tournament_size, replace=False)
fitness_scores = [self.fitness(self.decode(individual)) for individual in tournament]
return tournament[np.argmin(fitness_scores)]
def uniform_crossover(self, parent1, parent2):
mask = np.random.choice([0, 1], len(parent1))
return ''.join(p1 if m == 1 else p2 for p1, p2, m in zip(parent1, parent2, mask))
def mutate(self, individual):
return ''.join(bit if np.random.rand() > self.mutation_rate else np.random.choice(['0', '1'])
for bit in individual)
def run(self):
for generation in range(min(self.num_generations, 10)):
new_population = []
for _ in range(self.population_size):
parent1 = self.tournament_selection()
parent2 = self.tournament_selection()
child = self.uniform_crossover(parent1, parent2)
child = self.mutate(child)
new_population.append(child)
self.population = new_population
fitness_scores = [self.fitness(self.decode(individual)) for individual in self.population]
print(f'{generation + 1} generation: {fitness_scores}')
best_individual = self.population[np.argmin(fitness_scores)]
best_solution = self.decode(best_individual)
print(f'The roots are: [{best_solution}, {-best_solution}]')
# Constants
POPULATION_SIZE = 10
NUM_GENERATIONS = 10
MUTATION_RATE = 0.01
TOURNAMENT_SIZE = 5
# Run the Genetic Algorithm
ga = GeneticAlgorithm(POPULATION_SIZE, NUM_GENERATIONS,
MUTATION_RATE, TOURNAMENT_SIZE)
ga.run()
```
